```python
import numpy as np
from numpy.linalg import inv
import math
```

```python
def gaussian_kernel(x, y, cov):
    return math.exp(-0.5*float(np.transpose(x - y) * inv(cov) * (x - y)))
```

```python
x = np.matrix("1;2;3")
y = np.matrix("2;3;4")
cov = np.matrix("1,0,0;0,1,0;0,0,1")
```

```python
gaussian_kernel(x, y, cov)
```

```python
def polynomial(x, y, c, p):
    return float((c + np.transpose(x) * y)**p)
```

```python
polynomial(x, y, 1, 2)
```

```python
def linear(x, y):
    return float(np.transpose(x) * y)
```

```python
linear(x, y)
```
